```python
import pandas as pd
import json
import numpy as np
```

```python
sample_50_videos = pd.read_csv('sample_50_videos.csv')
with open('vslnet_9_3230_preds.json') as predictions_file:
    predictions_dict = json.load(predictions_file)
```

```python
predictions_mean = {
    'clip_start_sec_pred': [],
    'clip_end_sec_pred': []
}

for annotation, query_idx in zip(sample_50_videos['annotation_uid'], sample_50_videos['query_idx']):
    for sample_predicted in predictions_dict['results']:
        if sample_predicted['annotation_uid'] == annotation and sample_predicted['query_idx'] == query_idx:
            clip_start_sec_pred_mean, clip_end_sec_pred_mean = np.array(sample_predicted['predicted_times']).mean(axis=0)
            predictions_mean['clip_start_sec_pred'].append(clip_start_sec_pred_mean)
            predictions_mean['clip_end_sec_pred'].append(clip_end_sec_pred_mean)
```

```python
sample_50_videos_concat = pd.concat([sample_50_videos, pd.DataFrame(predictions_mean)], axis=1, ignore_index=True)
sample_50_videos_concat.columns = list(sample_50_videos.columns) + ['clip_start_sec_pred', 'clip_end_sec_pred']
sample_50_videos_concat.set_index("id")
```
